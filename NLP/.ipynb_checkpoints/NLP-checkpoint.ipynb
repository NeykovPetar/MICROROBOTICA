# Esercizio

leggere il file, creare un dizionario python che abbia come chiave ogni parola in minuscolo e come frequenza il numero di volte che compare nel testo

```python
def rimuoviPunteggiatura(testo,caratteri_da_eliminare):
    for carattere in caratteri_da_eliminare:
        testo = testo.replace(carattere," ")
    return testo
```

```python
def associaParole(dizionario,lista_parole):
    dizionario_previsioni = {}
    
    for i,parola in enumerate(lista_parole[:-1]):
        
        if parola not in dizionario_previsioni:
            dizionario_previsioni[parola] = []
            
        else:
            dizionario_previsioni[parola].append(lista_parole[i+1])
        
    return dizionario_previsioni
```

```python
file = open("C:/Users/Petar Neykov/Desktop/Scuola/MICROROBOTICA/LAB/NLP/testo.txt", "r")
caratteri_da_eliminare = ["," , ";" , "." , ":" , "-" , "_" , "#" , "*" , "/" , "|","»", "(" , ")", "$", '"',"!", "?","^","£","&", "%", "=", "{", "}","[", "]","¬","¨","«","'","©"]
testo = file.readlines()
testo = [rimuoviPunteggiatura(riga,caratteri_da_eliminare) for riga in testo]
testo = [riga[:-1].split(" ") for riga in testo]
file.close()
```

```python
dizionario = {}
```

```python
lista_parole = []
for riga in testo:
    for parola in riga:
        if parola != '':
            lista_parole.append(parola.lower())

#scrive in maniera più coincisa
```

```python
for parola in lista_parole:
    if parola not in dizionario:
        dizionario[parola] = 1
    else:
        dizionario[parola] += 1
```

```python
#dizionario
associaParole(dizionario,lista_parole)
```
